```python
# Importing standar modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# EDA of Final DataFrame

```python
# Importing data from csv file and creating a dataframe
final_year_2000 = pd.read_csv(r'Data\final_tmdb_data_2000.csv.gz', low_memory=False)
# Checking the first 5 rows of the dataframe
final_year_2000.head()
```

- I will be checking the csv files to ensure that the data files are properly imported into this notebook.

```python
# Importing data from csv file and creating a dataframe
final_year_2001 = pd.read_csv(r'Data\final_tmdb_data_2001.csv.gz', low_memory=False)
# Checking the first 5 rows of the dataframe
final_year_2001.head()
```

```python
# Concatenating both dataframes imported in the previous steps
final_df = pd.concat([final_year_2000, final_year_2001])
# Checking the first 5 rows of the dataframe
final_df.head()
```

Now I will be answering 

1. How many movies had at least some valid financial information (values > 0 for budget OR revenue)?


```python
## Creating filter to remove rows with 0 values in budget and revenue columns
filter = final_df['budget'] > 0
filter2 = final_df['revenue'] > 0
# Creating a new dataframe with the filtered data
plot_df = final_df[filter & filter2]
# Setting the style for the plots and creating the figure and axes
with plt.style.context(['fivethirtyeight', 'seaborn-muted']):
    fig, ax = plt.subplots(2, 1, figsize=(15, 15))
    # Using seaborn to plot the distribution of the budget and revenue columns
    sns.kdeplot(plot_df['budget'], ax=ax[0], color='Blue', shade=True, fill=True)
    # Setting the title, labels and ticks for the plot
    ax[0].set_title('Budget Distribution (in Millions)')
    ax[0].set_xlabel('Budget (in Millions)')
    ax[0].set_ylabel('')
    # Formatting the xticks to show the values in millions
    ax[0].set_xticklabels(['${:,.0f}'.format(x) for x in ax[0].get_xticks().tolist()], rotation=45, ha='right', fontsize=12)
    # Removing the yticks from the plot to avoid confusion
    ax[0].set_yticks([])
    # Using seaborn to plot the distribution of the budget and revenue columns
    sns.kdeplot(plot_df['revenue'], ax=ax[1], color='Green', shade=True, fill=True)
    # Setting the title, labels and ticks for the plot
    ax[1].set_title('Revenue Distribution (in Millions)')
    ax[1].set_xlabel('Revenue (in Millions)')
    ax[1].set_ylabel('')
    # Formatting the xticks to show the values in millions
    ax[1].set_xticklabels(['${:,.0f}'.format(x) for x in ax[1].get_xticks().tolist()], rotation=45, ha='right', fontsize=12)
    # Removing the yticks from the plot to avoid confusion
    ax[1].set_yticks([])
    # Setting the title for the figure
    fig.suptitle('Budget and Revenue Distribution', fontsize=20)
    # Setting the spacing between the plots
    plt.tight_layout()
    # Showing the plot
    plt.show();
```

2. How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
plot_df = final_df.apply(lambda row: row[final_df['certification'].isin(['G', 'PG', 'PG-13', 'R'])])
```

```python

with plt.style.context(['fivethirtyeight', 'seaborn-poster', 'seaborn-muted', 'seaborn-colorblind']):
    sns.countplot(x='certification', data=plot_df, order=plot_df['certification'].value_counts().index.sort_values())
    plt.title('Certification Distribution of Movies')
    plt.xlabel('Certifications')
    plt.ylabel('Counts of Movies')
    plt.xticks(fontsize=20)
    plt.show();
```

3 & 4: What is the average revenue per certification category? What is the average budget per certification category?

```python
plot_df = plot_df.groupby('certification').mean().reset_index()
with plt.style.context(['fivethirtyeight', 'seaborn-poster', 'seaborn-muted', 'seaborn-colorblind']):
    fig, ax = plt.subplots(1, 2, figsize=(15, 10))
    sns.barplot(x='certification', y='budget', data=plot_df, order=plot_df['certification'].value_counts().index.sort_values(ascending=False), ax=ax[0])
    ax[0].set_title('Average Budget by Certification')
    ax[0].set_xlabel('Certification')
    ax[0].set_ylabel('Average Budget (in Millions)')
    ax[0].set_yticklabels(['${:,.0f}'.format(x) for x in ax[0].get_yticks().tolist()], ha='right', fontsize=12)
    ax[0].set_ylim(0, max(plot_df['budget'])*1.1)
    sns.barplot(x='certification', y='revenue', data=plot_df, order=plot_df['certification'].value_counts().index.sort_values(ascending=False), ax=ax[1])
    ax[1].set_title('Average Revenue by Certification')
    ax[1].set_xlabel('Certification')
    ax[1].set_ylabel('Average Revenue (in Millions)')
    ax[1].set_yticklabels(['${:,.0f}'.format(x) for x in ax[1].get_yticks().tolist()], ha='right', fontsize=12)
    ax[1].set_ylim(0, max(plot_df['revenue'])*1.1)
    plt.tight_layout()
    plt.show();
```
